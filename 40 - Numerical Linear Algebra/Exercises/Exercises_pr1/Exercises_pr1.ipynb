**1.1 - Inizialization**

```python
import numpy as np

n = 4

A = np.random.random((n, n))
B = np.random.random((n, n))
print("A: ") 
print(A)
print("B: ")
print(B)
```

**1.2 - ijk Cycle** <br/>
*This Cylce computes in the first iteration of k indices the dot product of the i row of A with the j column of B. Then it proceeds with computing the same for the row i, changing the value of the column j. Finally, iterating i, it does the same on the remaining rows*

```python
C = np.zeros((n,n))
def ijk(A,B,C):
  for i in range(0,n):
    for j in range(0,n):
      for k in range(0,n):
        C[i,j] = A[i,k]*B[k,j] + C[i,j] 
  return C

C = ijk(A,B,C)
print(C)
if ((C-np.dot(A,B)).all()<10**-6):
  print("CHECK: TRUE")
else:
  print("CHECK: FALSE")
```

**1.2 - ijk Cycle (Optimized)**

```python
C = np.zeros((n,n))
def ijk(A,B,C):
  for i in range(0,n):
    for j in range(0,n):
      C[i,j] = np.dot(A[i,:],B[:,j]) + C[i,j] 
  return C

C = ijk(A,B,C)
print(C)
if ((C-np.dot(A,B)).all()<10**-6):
  print("CHECK: TRUE")
else:
  print("CHECK: FALSE")
```

**1.3 - jik Cycle** <br/>
*This cycle, similarly to the previous, computes the dot product of the i row of A and the j column of B. Then it proceeds iterating i, meaning computing the same k cycle for the first column of C. Finally, in the j iterations, it does the same on the following columns of C*

```python
C = np.zeros((n,n))
def jik(A,B,C):
  for j in range(0,n):
    for i in range(0,n):
      for k in range(0,n):
        C[i,j] = A[i,k]*B[k,j] + C[i,j] 
  return C

C = jik(A,B,C)
print(C)
if ((C-np.dot(A,B)).all()<10**-6):
  print("CHECK: TRUE")
else:
  print("CHECK: FALSE")
```

**1.4 - ikj Cycle** <br/>
*The first iterations of j compute the product between the i,k value of A and the row k of B, computing the first term of the dot product between the k column of A and the k column of B. It then proceeds iterating k, computing the following terms of the dot product. Finally it does the same for all rows i.*

```python
C = np.zeros((n,n))
def ikj(A,B,C):
  for i in range(0,n):
    for k in range(0,n):
      for j in range(0,n):
        C[i,j] = A[i,k]*B[k,j] + C[i,j] 
  return C

C = ikj(A,B,C)
print(C)
if ((C-np.dot(A,B)).all()<10**-6):
  print("CHECK: TRUE")
else:
  print("CHECK: FALSE")
```

**1.4 - ikj Cycle (Optimized)**

```python
C = np.zeros((n,n))
def ikj(A,B,C):
  for i in range(0,n):
    for k in range(0,n):
      aux = A[i,k]
      for j in range(0,n):
        C[i,j] = aux*B[k,j] + C[i,j] 
  return C

C = ikj(A,B,C)
print(C)
if ((C-np.dot(A,B)).all()<10**-6):
  print("CHECK: TRUE")
else:
  print("CHECK: FALSE")
```

**1.5 - jki Cycle (Optimized)**

```python
C = np.zeros((n,n))
def jki(A,B,C):
  for j in range(0,n):
    for k in range(0,n):
      aux=B[k,j]
      for i in range(0,n):
        C[i,j] = A[i,k]*aux + C[i,j] 
  return C

C = jki(A,B,C)
print(C)
if ((C-np.dot(A,B)).all()<10**-6):
  print("CHECK: TRUE")
else:
  print("CHECK: FALSE")
```

**1.6 - kij Cycle** <br/>
*The iteration of j computes all the first terms of the dot product for C(i,j) on the row i. The iteration of i does the same for the following rows. Finally, iterating k, adds all the other terms of the dot products in all the postions i,j of C*

```python
C = np.zeros((n,n))
def kij(A,B,C):
  for k in range(0,n):
    for i in range(0,n):
      for j in range(0,n):
        C[i,j] = A[i,k]*B[k,j] + C[i,j] 
  return C

C = kij(A,B,C)
print(C)
if ((C-np.dot(A,B)).all()<10**-6):
  print("CHECK: TRUE")
else:
  print("CHECK: FALSE")
```
