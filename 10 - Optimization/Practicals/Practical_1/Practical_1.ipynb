```python
#Function f_A ------------------------------------------------------------------

fig = plt.figure()
ax = fig.gca(projection='3d')

# Generate the points
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
f_A = -X**2-Y**2

# Plot the surface
surf = ax.plot_surface(X, Y, f_A, cmap=cm.coolwarm, alpha=0.5,
                       linewidth=0, antialiased=False)
plt.title("f_A(x)")    

#Function f_B ------------------------------------------------------------------

fig = plt.figure()
ax = fig.gca(projection='3d')

# Generate the points
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
f_B = X**2-Y**2

# Plot the surface
surf = ax.plot_surface(X, Y, f_B, cmap=cm.coolwarm, alpha=0.5,
                       linewidth=0, antialiased=False)
plt.title("f_B(x)")    

#Function f_C ------------------------------------------------------------------

fig = plt.figure()
ax = fig.gca(projection='3d')

# Generate the points
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
f_C = X**2

# Plot the surface
surf = ax.plot_surface(X, Y, f_C, cmap=cm.coolwarm, alpha=0.5,
                       linewidth=0, antialiased=False)
plt.title("f_C(x)")    
```

**2.1.4 (b) - Evaluating the Hessian at the stationary points**

```python
from numpy import linalg as LA

#f_A
print("Hessian of f_A: ")
H_A = np.array([[-2, 0], [0, -2]])
print(H_A)
eigvals_A = LA.eigvals(H_A)
print("eigenvalues of A: " + str(eigvals_A) + "\n")
print("Both the eigenvalues of the Hessian of f_A at the stationary point are negative, which means that the Hessian at that point is negative definite")
print("Looking at the Taylor expansion of order two of the function, we can conclude that the point is (0,0) is a maximum \n")

#f_B
print("Hessian of f_B: ")
H_B = np.array([[2, 0], [0, -2]])
print(H_B)
eigvals_B = LA.eigvals(H_B)
print("eigenvalues of B: " + str(eigvals_B) + "\n")
print("The eigenvalues of the Hessian of f_A at the stationary point are one positive and one negative, which means that the Hessian at that point is neither negative nor positive definite")
print("Looking at the Taylor expansion of order two of the function, we cannot classify the point (0,0) as a maximum or a minimum")
print("However, we can conclude that the function is increasing in one direction and decreasing in the other, which means that the point (0,0) is a saddle \n")


#f_C
print("Hessian of f_C: ")
H_C = np.array([[2, 0], [0, 0]])
print(H_C)
eigvals_C = LA.eigvals(H_C)
print("eigenvalues of C: " + str(eigvals_C) + "\n")
print("The eigenvalues of the Hessian of f_A at the stationary point are one positive and the other null")
print("This means that in one direction the function is minimized at (0,0), while in the other there is no variation even at order two (as well as on order one, as previously noted)")
print("Looking at the Taylor expansion of order two of the function, we can conclude that the points on the line going through (0,0) with direction v=(1,0) are minima")
```

**2.2.1 - Plot the function in the range [-2,2]x[-1,1] with a step of 0.1**

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')

# Generate the points
X1 = np.arange(-2, 2, 0.1)
X2 = np.arange(-1, 1, 0.1)
X1, X2 = np.meshgrid(X1, X2)
f = (X1**2)*(4-2.1*(X1**2)+1/3*(X1**4))+X1*X2+(X2**2)*(-4+4*(X2**2))

# Plot the surface
surf = ax.plot_surface(X1, X2, f, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.title("f(x)")    
```

```python
# Contour plot of f(x)
fig = plt.figure()

plt.contourf(X1, X2, f, 20, cmap='RdGy')
plt.colorbar()
```

**2.2.3 - Evaluating the norm of the gradient of f(x) squared to approximate stationary points**

```python
# Gradient of f analitically computed: G_f = (8*x1-8.4*x1^3+2*x1^5+x2, x1-8*x2+16*x2^3)
# a) Norm squared of the gradient N_G: creates a matrix for the values of x1 on the interval [-2,2] with a step of 0.005 on the rows 
# and the values of x2 on the interval [-1,1] with a step of 0.005 on the columns
X1 = np.arange(-2, 2, 0.005)
X2 = np.arange(-1, 1, 0.005)
n_x1 = (2-(-2))/0.005
n_x2 = (1-(-1))/0.005
N_G = np.zeros([int(n_x1),int(n_x2)])
N_G.shape
```

```python
# Gradient of f analitically computed: G_f = (8*x1-8.4*x1^3+2*x1^5+x2, x1-8*x2+16*x2^3)
# a) Norm squared of the gradient N_G: creates a matrix for the values of x1 on the interval [-2,2] with a step of 0.005 on i rows 
# and the values of x2 on the interval [-1,1] with a step of 0.005 on j columns
X1 = np.arange(-2, 2, 0.005)
X2 = np.arange(-1, 1, 0.005)
n_x1 = (2-(-2))/0.005
n_x2 = (1-(-1))/0.005
N_G = np.zeros([int(n_x1),int(n_x2)])
for i in range(0, int(n_x1)):    
  for j in range(0, int(n_x2)): 
    N_G[i,j] = (8*X1[i]-8.4*(X1[i]**3)+2*(X1[i]**5)+X2[j])**2+(X1[i]-8*X2[j]+16*(X2[j]**3))**2

# b)c) Search for the values in of the gradient close to 0 to find potential candidates for minimum, maximum and saddle points
Stat_points_i = []
Stat_points_j = []
for i in range(1, int(n_x1-1)):      # <- not checking the border of the matrix, values of x1
  for j in range(1, int(n_x2-1)):      # <- not checking the border of the matrix, values of x2
    if (N_G[i,j]<N_G[i+1,j+1] and N_G[i,j]<N_G[i,j+1] and N_G[i,j]<N_G[i-1,j+1] and N_G[i,j]<N_G[i-1,j] and N_G[i,j]<N_G[i-1,j-1] and N_G[i,j]<N_G[i,j-1] and N_G[i,j]<N_G[i+1,j-1] and N_G[i,j]<N_G[i+1,j]):
      Stat_points_i.append(i)
      Stat_points_j.append(j)
      print("Norm_G close to zero in: " + str(i) + "," + str(j) + ", corresponding to the point: " + str(X1[i]) + "," + str(X2[j]) + " Norm_G=" + str(N_G[i,j]))
```

```python
print(Stat_points_i)     # These two arrays contain the indices of the vectors X1 and X2 at whcich we found stationary points, meaning points with the norm of the gradient close to null
print(Stat_points_j)
for i in range(0, len(Stat_points_i)):      # Stationary Points
  print("(" + str(X1[Stat_points_i[i]]) + "," + str(X2[Stat_points_j[i]]) + ")")
```

```python
# Show the results with better visualization: we can observe how the value of the norm of the gradient squared 
# in the identified points is close to zero, as we were expecting doing this "brute force" search

import pandas as pd

data = np.zeros([len(Stat_points_i), 5])

data[:,0] = Stat_points_i
data[:,1] = Stat_points_j
data[:,2] = X1[Stat_points_i]
data[:,3] = X2[Stat_points_j]
data[:,4] = N_G[Stat_points_i, Stat_points_j]

df = pd.DataFrame(data)
df.columns = ['i index', 'j index', 'x', 'y', 'Norm^2 of the gradient']
df
```

**2.2.4 - Evaluating the Hessian of the function at the stationary points**

```python
from numpy import linalg as LA

def Hessian_mat(x1,x2):
  H_f = np.matrix([[10*(x1**4)-25.2*(x1**2)+8, 1], [1, 48*(x2**2)-8]])
  return H_f

Minimums_x1 = []
Minimums_x2 = []
classification = []
lambdas1 = []
lambdas2 = []
for i in range(0, len(Stat_points_i)):
  H_f = Hessian_mat(X1[Stat_points_i[i]],X2[Stat_points_j[i]])
  [lambda1, lambda2] = LA.eigvals(H_f)
  lambdas1.append(lambda1)              # These lines are reduntant, the reason is they were added later 
  lambdas2.append(lambda2)              # in order to visualize the data better in the following section
  if lambda1>0 and lambda2>0:
    print("The point " + str(X1[Stat_points_i[i]]) + str(X2[Stat_points_j[i]]) + "is a minimum, with: lambda1= " + str(lambda1) + " lambda2= " + str(lambda2))
    # The Hessian of the function computed in that point is positive definite
    Minimums_x1.append(X1[Stat_points_i[i]])
    Minimums_x2.append(X2[Stat_points_j[i]])
    classification.append('Minimum')
  elif lambda1<0 and lambda2<0:
    print("The point " + str(X1[Stat_points_i[i]]) + str(X2[Stat_points_j[i]]) + "is a maximum, with: lambda1= " + str(lambda1) + " lambda2= " + str(lambda2))
    classification.append('Maximum')
    # The Hessian of the function computed in that point is negative definite
  else:
    print("The point " + str(X1[Stat_points_i[i]]) + str(X2[Stat_points_j[i]]) + "is a saddle, with: lambda1= " + str(lambda1) + " lambda2= " + str(lambda2))
    classification.append('Saddle')
    # The Hessian of the function computed in that point is neither positive definite nor negative definite
```

```python
# Show the results with better visualization

import pandas as pd

data = np.zeros([len(Stat_points_i), 5]).astype(object)

data[:,0] = X1[Stat_points_i]
data[:,1] = X2[Stat_points_j]
data[:,2] = classification
data[:,3] = lambdas1
data[:,4] = lambdas2

df = pd.DataFrame(data)
df.columns = ['x', 'y', 'Classification of the point', 'lambda1', 'lambda2']
df
```
