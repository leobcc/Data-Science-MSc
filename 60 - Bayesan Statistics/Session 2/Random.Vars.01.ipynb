## Gamma r.v.

### Gamma function

$$
	\Gamma(t)=\int_{0}^{\infty} x^{t-1} e^{-x}\,dx.
$$

For integer $a$, $\Gamma(a)=(a-1)!$. In general, $\Gamma(t+1)=t\cdot\Gamma(t)$.

$$
    \Gamma\left(\tfrac{1}{2}\right)=\sqrt{\pi}.
$$

### Gamma distribution

A r.v. $X$ has a $\operatorname{Gamma}(\alpha,\beta)$ distribution with _shape_ parameter $\alpha>0$ and _rate_ parameter $\beta>0$ if it is absolutely continuous on $(0,\infty)$ with pdf:

$$
    f(x|\alpha,\beta)=\dfrac{\beta^{\alpha}}{\Gamma(\alpha)}\,x^{\alpha-1}\,\exp\{-\beta\,x\},
    \mskip20mu x>0.
$$

In particular, for $\alpha=1$ $\operatorname{Gamma}(1,\beta)$ is the exponential $\operatorname{Exp}(\beta)$ distribution.

For $X\sim\operatorname{Gamma}(\alpha,\beta)$,

$$
    \operatorname{E}(X)=\dfrac{\alpha}{\beta},\mskip40mu
    \operatorname{var}(X)=\dfrac{\alpha}{\beta^{2}}.
$$

##### Remark:

_Read the [Wikipedia article about the gamma distribution.](https://en.wikipedia.org/wiki/Gamma_distribution)_ 

Similarly with the distributions below.

## Beta r.v.

### Beta function
   
Definition:

$$
    B(r,s)=\int_{0}^{1}t^{r-1}\,(1-t)^{s-1}\,dt
$$


This function relates to the Gamma function through the identity:

$$
    B(r,s)=\frac{\Gamma(r)\,\Gamma(s)}{\Gamma(r+s)}.
$$

Useful identities:

$$
    \begin{array}{lcl}
    B(r,s)		&=&B(s,r),\\
    B(r,s)		&=&B(r,s+1)+B(r+1,s),\\
    B(r+1,s) 	&=&B(r,s)\cdot\dfrac{r}{r+s},\\
    B(r,s+1) 	&=&B(r,s)\cdot\dfrac{s}{r+s}. 
    \end{array}
$$

For integer $r$, $s$,

$$
    B(r,s)=\frac{(r-1)!\,(s-1)!}{(r+s-1)!}=\dfrac{1}{(r+s-1)\times\displaystyle{r+s-2\choose r-1}}.\\
$$

Equivalently,

$$
    {n\choose k}= \frac{1}{(n+1)\,B(n-k+1,k+1)}. 
$$

### Beta distribution

A r.v. $X$ with support on $[0,1]$ follows a $\operatorname{Beta}$ distributions with parameters $\alpha,\beta>0$, $X\sim\operatorname{Beta}(\alpha,\beta)$, if it is absolutely  continuous, with pdf:

$$
   f(x\,|\,\alpha,\beta)=\frac{1}{B(\alpha,\beta)}\,x^{\alpha-1}\,(1-x)^{\beta-1}.
$$

$U$ has expectation:

$$
    \operatorname{E}(X)=\frac{\alpha}{\alpha+\beta},
$$

and variance:

$$
    \operatorname{var}(X)=\frac{\alpha\,\beta}{(\alpha+\beta)^{2}\,(\alpha+\beta+1)}.
$$
	
Alternative parameterization, with $(\theta,p)$, defined by:

$$
	{\renewcommand{\arraystretch}{1.5}\begin{array}{ccc}
		\left\{\begin{array}{lcl}
		\theta	&=	&\alpha+\beta,\\[0.2cm]
		p		&=	&\dfrac{\alpha}{\alpha+\beta}.
		\end{array}\right.
		&\mskip20mu
		&\left\{\begin{array}{lcl}
		\alpha	&=	&\theta\cdot p,\\[0.2cm]
		\beta	&=	&\theta\cdot (1-p).
		\end{array}\right.
	\end{array}}
$$

With these parameters:

$$
	\operatorname{E}(X)=p,\mskip30mu \operatorname{var}(X)=\dfrac{p\cdot(1-p)}{\theta+1}.
$$

```r
old.par<-par(mfrow=c(3,3))
options(repr.plot.width=15,repr.plot.height=15)
u<-seq(0,1,length=10000)
for (a in 1:3){
    for (b in 1:3){
        x<-dbeta(u,a,b)
        header<-sprintf("Beta(%2.1f,%2.1f) pdf",a,b)
        plot(u,x,"l",col="blue",lwd=2.0,xlab="",ylab="",main=header,cex.main=1.8)
        }
    }
par(old.par)
```

## Bell-shaped pdf's: Normal, Cauchy, Student's $t$, Logistic, Laplace

### Student's $t$ distribution

A r.v. $X$, with support on $\mathbb{R}$ follows the  _Student's_ $t$ distribution with $\nu>0$ _degrees of freedom,_ _location parameter_ $\mu\in\mathbb{R}$ and scale parameter $\sigma>0$, $t(\nu,\mu,\sigma)$, if it is absolutely continuous, with pdf:

$$
    f(x)=\dfrac{\Gamma\left(\dfrac{\nu+1}{2}\right)}
    {\Gamma\left(\dfrac{\nu}{2}\right)\,\sqrt{\pi\,\nu}\,\sigma}\,
    \left(1+\dfrac{1}{\nu}\left(\dfrac{x-\mu}{\sigma}\right)^{2}\right)^{-\frac{\nu+1}{2}}
    ,\mskip30mu x\in\mathbb{R}.
$$

This $X$ is the result of applying a translation by $\mu$ and scaling by $\sigma$ to a _standard Student's_ $t$ r.v. with $\nu>0$ _degrees of freedom,_  $t(\nu)$, whose pdf is that from above with $\mu=0$ and $\sigma=1$. 

$$
    f(x)=\dfrac{\Gamma\left(\dfrac{\nu+1}{2}\right)}
    {\Gamma\left(\dfrac{\nu}{2}\right)\,\sqrt{\pi\,\nu}}\,
    \left(1+\dfrac{x^{2}}{\nu}\right)^{-\frac{\nu+1}{2}}
    ,\mskip30mu x\in\mathbb{R}.
$$


As its name and origin suggests, usually $\nu\in\mathbb{N}$.

_Cauchy distribution_ is another name for $t(1)$, Student's $t$ with 1 degree of freedom.

#### Caution

Moments of order $k$, $\operatorname{E}(X^{k})$ of $X\sim t(\nu)$ exist only for $k<\nu$. In particular, a Cauchy r.v. has neither expectation (nor variance, etc). For $X\sim t(\nu)$ variance exists only when $\nu>2$ and then:

$$
    \operatorname{E}(X)=0,\mskip40mu
    \operatorname{var}(X)=\dfrac{\nu}{\nu-2}.
$$

This property is associated with the fact that, for small $\nu$, this family of distributions has _heavy tails,_ meaning that there is a significant probability of registering values very far from its geometric center. As $\nu$ increases$t(\nu)\rightarrow\operatorname{N}(0,1)$, Students's $t$ distribution approaches the normal.

```r
options(repr.plot.width=9,repr.plot.height=9)
u<-seq(-6,6,length=1000)
plot(u,dnorm(u),type="l",lwd=3.0,col="red",xlab="x",ylab="Density",ylim=c(0,0.45),
    main="Comparing standard normal and Student's t pdf's")
lines(u,dt(u,df=1),lwd=2.5,col="blue")
lines(u,dt(u,df=2),lwd=2.5,col="cyan")
legend("topright",c("N(0,1)","t(1)","t(2)"),lwd=2.5,col=c("red","blue","cyan"))
```

### Logistic distribution

A r.v. $X$, with support on $\mathbb{R}$ follows the _Logistic_ distribution with location parameter $\mu\in\mathbb{R}$ and scale parameter $\sigma>0$ if it is absolutely continuous, with pdf:

$$
    f(x)=\dfrac{1}{\sigma}\,
    \dfrac{\exp\left(-\dfrac{x-\mu}{\sigma}\right)}
    {1+\exp\left(-\dfrac{x-\mu}{\sigma}\right)}
    =\dfrac{1}{4\,\sigma}\,
    \operatorname{sech}^{2}\left(\dfrac{x-\mu}{2\,\sigma}\right),\mskip30mu x\in\mathbb{R}.
$$

Its cdf is the _Logistic function,_ (translated by $\mu$ and scaled by $\sigma$).

If $X\sim\operatorname{Logistic}(\mu,\sigma)$,

$$
\operatorname{E}(X)=\mu,\mskip40mu \operatorname{var}(X)=\dfrac{\sigma^{2}\,\pi^{2}}{3}.
$$

```r
options(repr.plot.width=9,repr.plot.height=9)
u<-seq(-6,6,length=1000)
plot(u,dnorm(u),type="l",lwd=3.0,col="red",xlab="x",ylab="Density",ylim=c(0,0.45),
    main="Comparing standard normal and Logistic pdf's")
lines(u,dlogis(u),lwd=2.5,col="blue")
lines(u,dlogis(u,scale=sqrt(3)/pi),lwd=2.5,col="cyan")
legend("topright",c("N(0,1)","Logistic(s=1)",expression(paste("Logistic(s=",sqrt(3)/pi,")"))),
       lwd=2.5,col=c("red","blue","cyan"))
```

### Laplace distribution

A r.v. $X$, with support on $\mathbb{R}$ follows the _Laplace_ or _double exponential_ distribution with parameters $\mu\in\mathbb{R}$ and $\sigma>0$ if it is absolutely continuous, with pdf:

$$
    f(x)=\dfrac{1}{2\,\sigma}\,\exp\left(-\dfrac{|x-\mu|}{\sigma}\right),\mskip30mu x\in\mathbb{R}.
$$

If $X\sim\operatorname{Laplace}(\mu,\sigma)$,

$$
\operatorname{E}(X)=\mu,\mskip40mu \operatorname{var}(X)=2\,\sigma^{2}.
$$

# Expectation, variance, moments of a discrete random variable

```r
# An example discrete r.v.
x<-c(1,2,3,4)
d<-c(0.2, 0.3, 0.4, 0.1)
```

```r
EX<-sum(x*d)
EX
```

#### A function to compute the expectation of a discrete r.v.

```r
E<-function(x,d){
    EX<-sum(x*d)
    return(EX)
    }
```

## Variance of a random variable

```r
# deviations with respect to the expectation
D<-x-EX
D
```

```r
# Squared deviations
D2<-D^2
D2
```

```r
# Variance is the expectation of the squared deviations w.r.t. the expectation
VarX<-E(D2,d)
VarX
```

#### A function to compute the variance of a discrete r.v.

```r
Var<-function(x,d){
    EX<-E(x,d)
    D2<-(x-EX)^2
    VarX<-E(D2,d)
    return(VarX)
}
```

```r
Var(x,d)
```

## Moments of a discrete r.v.

For each positive integer $k$, the expectation of $X^k$.

For $k=1$, $M_1$ is the expectation.

```r
M<-function(x,d,k){
    Mk<-E(x^k,d)
    return(Mk)
}
```

```r
# First moment
M(x,d,1)
```

```r
# Second moment
MX2<-M(x,d,2)
MX2
```

### Variance in terms of moments

Identity, the variance is the difference between the second moment and the first moment squared:

$$
    \operatorname{var}(X)=\operatorname{E}(X^{2})-(\operatorname{E}(X))^2
$$

```r
round(M(x,d,2)-E(x,d)^2,4)
```

#### Example: expectation of a binomial r.v.

Repeat these computations with different parameters $n$ and $\theta$

```r
n<-25
theta<-0.35
x<-0:n # Values vector
d<-dbinom(x,size=n,prob=theta)
EX<-E(x,d)
EX
# Actually there is a closed formula
n*theta
```

# Expectation, variance, moments of a continuous random variable

### Expectation of a continuous r.v.  

If $X$ is a continuous r.v. with pdf $f$, we define:
	
$$
	\operatorname{E}(X)=\int_{-\infty}^{\infty}x\,f(x)\,dx,
$$

if this integral exists. 

### Expectation of a function of a continuous r.v. 

Similarly for a function $g(X)$ of a continuous r.v. $X$:

$$
    \operatorname{E}(g(X))=\int_{-\infty}^{\infty}g(x)\,f(x)\,dx.
$$

In particular, moments, variance.

For some easy cases it is possible to obtain expectations analytically. For most interesting cases, expectations will be obtained by:

(1) Approximation

(2) Simulation

#### Example: expectation of a gamma distribution

Actually the expectation of a gamma pdf has a closed formula expression. See Wikipedia.

The expectation of $X\sim\operatorname{Gamma}(\alpha,\beta)$ is: $\operatorname{E}(X)=\dfrac{\alpha}{\beta}$.

```r
# Example values
alpha<-4.5
beta<-2.46
```

```r
xm<-10 # A tentative maximum x value, depending on alpha and beta (pdf should be almost zero there)
x<-seq(0,xm,length=500)
options(repr.plot.width=7,repr.plot.height=7)
plot(x,dgamma(x,shape=alpha,rate=beta),type="l",lwd=2.5,col="blue",ylab="Density")
abline(h=0,col="black",lw=1.5)
```

```r
# Expectation by numerical integration
xf<-function(x,...){x*dgamma(x,...)}
EX<-integrate(xf,lower=0,upper=xm,shape=alpha,rate=beta)$value
round(EX,6)
```

```r
# Expectation from the closed formula expression:
round(alpha/beta,6)
```

```r
options(repr.plot.width=7,repr.plot.height=7)
plot(x,dgamma(x,shape=alpha,rate=beta),type="l",lwd=2.5,col="blue",ylab="Density")
abline(h=0,col="black",lw=1.5)
lines(c(EX,EX),c(0,dgamma(EX,shape=alpha,rate=beta)),lwd=2.5,col="LightBlue")
points(c(EX,EX),c(0,dgamma(EX,shape=alpha,rate=beta)),pch=19,cex=1.2,col="blue")
```
